```python
def compute_mean(list):
    mean = sum(list) / len(list)
    return mean

list = ([2, 4, 6, 8])
compute_mean(list)
```

# Code explanation
The function above takes a list, sums the elements in a list, and then divide that sum by the total number of elements in the list to return the mean.

$\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i$
